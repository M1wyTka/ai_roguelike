```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def patrol(hp, dist):
    return 50  + np.random.randn() * 0 

def attack2(base_dist, allies_num):
  return 200*(3 - base_dist);

def go_base(base_dist, allies_num):
    return 150*(1-allies_num) + base_dist*10 + np.random.randn() * 0
```

```python
ALLIES = np.arange(0, 5, 0.25)
DIST = np.arange(0, 10, 0.025)
img = np.zeros((ALLIES.shape[0], DIST.shape[0]))
ht = np.zeros((3, ALLIES.shape[0], DIST.shape[0]))
# TODO: Vectorize
for x, ally in enumerate(ALLIES):
    for y, ds in enumerate(DIST):
        utility = np.array([attack2(ds, ally), go_base(ds, ally), patrol(ally, ds)])
        for i in range(utility.shape[0]):
            ht[i][x][y] = utility[i]
        img[x][y] = np.argmax(utility)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(img, extent=[0, 10, 5, 0])
ax.set_aspect('auto')
ax.set_xlabel("dist")
ax.set_ylabel("ally")


plt.colorbar(im)

fig, ax = plt.subplots(2, 2)
for i in range(3):
    im = ax[i %2, i // 2].imshow(ht[i], extent=[0, 10, 5, 0], cmap='Greys')
    ax[i %2, i // 2].set_aspect('auto')
    ax[i %2, i // 2].set_xlabel("dist")
    ax[i %2, i // 2].set_ylabel("ally")
    plt.colorbar(im)
```
